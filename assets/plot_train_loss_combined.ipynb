```python
import numpy as np
import matplotlib.pyplot as plt
import torch

n_epochs = 80

for model_name in ["s4d", "rnn", "gru", "lstm"]:
    checkpoint = torch.load(f'../checkpoint/{model_name}_ckpt.pth')
    losses = checkpoint['loss']
    epoch = checkpoint['epoch']

    xaxis = np.arange(1, n_epochs + 1)
    plt.plot(xaxis, losses, label=f'Training Loss for {model_name}')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.yscale('log')
plt.title('Comparison of RNN, GRU, LSTM and S4D on WMT2014 English-French')
plt.legend(fontsize="large")
plt.tight_layout()
plt.savefig("./train_loss_combined.png")
plt.savefig("./train_loss_combined.pdf")
plt.show()
```
